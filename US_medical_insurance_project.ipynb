# U.S. Medical Insurance Costs

```python
import csv

#CSV file columns into lists
age = []
sex = []
bmi = []
children = []
smoker_status = []
region = []
charges = []
```

```python
def load_list_data(lst, csv_file, column_name):
    
    with open(csv_file) as csv_info:
        csv_dict = csv.DictReader(csv_info)
        for row in csv_dict:
            lst.append(row[column_name])
        return lst
```

```python
load_list_data(age, 'insurance.csv', 'age')
load_list_data(sex, 'insurance.csv', 'sex')
load_list_data(bmi, 'insurance.csv', 'bmi')
load_list_data(children, 'insurance.csv', 'children')
load_list_data(smoker_status, 'insurance.csv', 'smoker')
load_list_data(region, 'insurance.csv', 'region')
load_list_data(charges, 'insurance.csv', 'charges')
```

**Function to calculate average**

```python
def average_calculation(list):
    total_list = 0
    for item in list:
        total_list += float(item)
    
    average = total_list / len(list)
    return average
```

**Analysis of average age**

```python
average_age = average_calculation(age)
rounded_age = round(average_age, 1)
print(f'The average age of the patients is: {rounded_age} years.')
```

**Count of males vs females**

```python

males = 0
females = 0

for individual in sex:
    if individual == 'male':
            males += 1
    elif individual == 'female':
            females += 1

print(f'There are: {males} male patients.')
print(f'There are: {females} female patients.')
```

**Patient regions**

```python
unique_regions = []

for place in region:
    if place not in unique_regions:
        unique_regions.append(place)

print(f'The unique regions are: {unique_regions}.')
```

**Average medical charges**

```python
average_insurance_charge = average_calculation(charges)
rounded_ins_average = round(average_insurance_charge, 2)
print(f'The rounded average of insurance charges of the patients is: ${rounded_ins_average}.')
```

**Smokers vs non-smokers**

```python
smokers = 0
non_smokers = 0

for individual in smoker_status:
    if individual == 'yes':
        smokers += 1
    elif individual == 'no':
        non_smokers += 1

print(f'{smokers} patients are smokers')
print('vs.')
print(f'{non_smokers} patients that are non-smokers')
```

**Difference in the insurance average of smoker patients vs non-smoker patients**

```python
zipped_cost_status = list(zip(smoker_status, charges))
print(zipped_cost_status)
```

```python

smoker_average = 0
non_average = 0
smoker_total = 0
non_total = 0

for item in zipped_cost_status:
    if item[0] == 'yes':
         smoker_total += float(item[1])

smoker_average = smoker_total / smokers


for item in zipped_cost_status:
    if item[0] == 'no':
        non_total += float(item[1])
    
non_average = non_total / non_smokers

rounded_smoker = round(smoker_average, 2)
rounded_non = round(non_average, 2)

print(f'The average insurance cost for smokers is: ${rounded_smoker}.')
print(f'The average insurance cost for non-smokers is: ${rounded_non}.')

difference = rounded_smoker - rounded_non

print(f'The difference in the average cost between smoker and non-smoker patient is: ${difference}.')
```

**Relation between BMI, smoking status, and sex**

```python
zip_list = list(zip(bmi, smoker_status, sex))
print(zip_list)
```

```python
totalbmi_male_smoker = 0
count_male_smoker = 0

totalbmi_male_non = 0
count_male_non = 0

totalbmi_female_smoker = 0
count_female_smoker = 0

totalbmi_female_non = 0
count_female_non = 0

for item in zip_list:
    if item[1] == 'yes' and item[2] == 'male':
        count_male_smoker += 1
        totalbmi_male_smoker += float(item[0])
    elif item[1] == 'no' and item [2] == 'male':
        count_male_non += 1
        totalbmi_male_non += float(item[0])
    elif item[1] == 'yes' and item[2] == 'female':
        count_female_smoker += 1
        totalbmi_female_smoker += float(item[0])
    elif item[1] == 'no' and item[2] == 'female':
        count_female_non += 1
        totalbmi_female_non += float(item[0])

male_bmi_smoker_avg = round((totalbmi_male_smoker / count_male_smoker), 1)
print(f'The average BMI for male smokers is: {male_bmi_smoker_avg}.')
male_bmi_non_avg = round((totalbmi_male_non / count_male_non), 1)
print(f'The average BMI for male non-smokers is: {male_bmi_non_avg}.')
female_bmi_smoker_avg = round((totalbmi_female_smoker / count_female_smoker), 1)
print(f'The average BMI for female smokers is: {female_bmi_smoker_avg}.')
female_bmi_non_avg = round((totalbmi_female_non / count_female_non), 1)
print(f'The average BMI for female non-smokers is: {female_bmi_non_avg}.')
```

**Region with the most smokers**

```python
region_smoker_zip = list(zip(smoker_status, region))
print(region_smoker_zip)
```

```python
southwest_smokers = 0
southeast_smokers = 0
northwest_smokers = 0
northeast_smokers = 0

for item in region_smoker_zip:
    if item[0] == 'yes' and item[1] == 'southwest':
        southwest_smokers += 1
    elif item[0] == 'yes' and item[1] == 'southeast':
        southeast_smokers += 1
    elif item[0] == 'yes' and item[1] == 'northwest':
        northwest_smokers += 1
    elif item[0] == 'yes' and item[1] == 'northeast':
        northeast_smokers += 1
    
def region_smoker(region, smokers):
    print(f'in the {region}, the number of patients that smoke is: {smokers}.')

region_smoker('Southwest', southwest_smokers)
region_smoker('Southeast', southeast_smokers)
region_smoker('Northwest', northwest_smokers)
region_smoker('Northeast', northeast_smokers)
print('\n')
print('The region with the most amount of smokers is the Southeast region.')
```
